### ***Tratando e filtrando os dados***

#### 2. Tratar os dados nulos

### ***Remover registros inconsistentes***

```python
# Remover registros que não fazem sentido ou que simplesmente não precisam constar na base de dados
# Apartamentos que possuem valor de aluguel igual a 0 e também condominio igual a 0
# Selecionar todas as linhas que desejamos remover
df.query('Valor == 0 | Condominio == 0')
```

```python
# Capturar os índices dessas linhas usando o .index
df.query('Valor == 0 | Condominio == 0').index
```

```python
# Armazenar todos esses índices em uma variável chamada de registros_a_remover
registros_a_remover = df.query('Valor == 0 | Condominio == 0').index
```

```python
# Remover as linhas que possuem esses índices da base de dados usando um método drop
df.drop(registros_a_remover, axis=0, inplace=True)
```

### ***Filtros***

```python
# Salvar essa seleção em uma segunda variável e passá-la para o dataframe
selecao2 = df['Valor'] < 1200
df[selecao2].head()
```

```python
# Aplicar ao dataframe
selecao_final = (selecao1) & (selecao2)
df[selecao_final].head()
```

```python
# Salvar essa seleção no dataframe em outra variável
df_1 = df[selecao_final]
```
